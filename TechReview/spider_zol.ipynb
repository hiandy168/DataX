```python
def fetch_scores(pid, product):
    for page in range(1, 100):
        url = cmt_ent.format(pid, page)
        print(url)
        res = req.get(url)
        lst = res.json()['list']
        comments = reCmt.findall(lst)
        if not comments:
            break
        scores = []
        for cmt in comments:
            tot = reTot.findall(cmt)
            fea = reFeature.findall(cmt)
            score = dict(fea)
            score['overall'] = tot and tot[0] or 'NA'
            score['product'] = product
            score['source'] = 'ZOL'
            score['pid'] = pid
            score['kind'] = 'Scores'
            
            scores.append(score)
        DB.insert_many(scores)
```
