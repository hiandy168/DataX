```python
import time, random
```

```python
def fetch(url, sleep=True):
    if sleep:
        time.sleep(random.random())
    print("Fetching {}".format(url))
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36',
    }
    
    res = req.get(url)
    return res
```

```python
API = 'https://movie.douban.com/j/new_search_subjects?sort=T&range=0,10&tags=%E7%94%B5%E5%BD%B1&start={}'
```

```python
for page in range(0, 5000, 20):
    api = API.format(page)
    
    res = fetch(api)
    
    DB.insert_many(res.json().get("data"))
```

```python
import re
```

```python
rePlst = re.compile(r'<li>[\s\S]*?<a class="playBtn" data-cn="(?P<site>[\s\S]*?)"[\s\S]*?<span class="buylink-price"><span>(?P<price>[\s\S]*?)</span></span>')
```

```python
reVotes = re.compile(r'<span property="v:votes">(\d+)<\/span>')
reDate = re.compile(r'<span property="v:initialReleaseDate" content="(\d{4}-\d{2}-\d{2}).*?">')
reScore = re.compile(r'property="v:average">(.*?)<\/strong>')
reTags = re.compile(r'<span property="v:genre">([\s\S]*?)<\/span>')
```

```python
movie = DB.find_one({"p":False})
url = movie.get("url")
html = fetch(url).text
```

```python
reVotes.findall(html)
reDate.findall(html)
reScore.findall(html)
reTags.findall(html)
```

```python
def df_lst(l, placeholder=None):
    if l:
        return l[0]
    else:
        return placeholder
```

```python
movie = DB.find_one({"p": False})
while movie:
    url = movie.get("url")
    
    res = fetch(url)
    html = res.text
    
    date = reDate.findall(html)
    votes = reVotes.findall(html)
    score = reScore.findall(html)
    tags = reTags.findall(html)
    
#     playlist = rePlst.finditer(res.text)
    
    """
    ps = []
    for p in playlist:
        site = p.groupdict()
        site['price'] = site['price'].strip()
        ps.append(site)
    """
    DB.update_one({"_id": movie.get("_id")}, {"$set": {"p": True,
                                                       "date": df_lst(date),
                                                       "voters": df_lst(votes, 0),
                                                       "score": df_lst(score, 0.0),
                                                       "tags": tags}})
    movie = DB.find_one({"p":False})
```
