```python
from db import DB
```

```python
import pandas as pd
```

```python
movies = DB.find({})
```

```python
vPD = {
    'site': [],
    'price': [],
    'movie_name': [],
    'movie_date': [],
    'movie_voters': [],
    'movie_score': [],
    'movie_id': [],
    'movie_tags': []
}
```

```python
for movie in movies:
    for play in movie.get("playlist"):
        vPD['site'].append(play.get("site"))
        vPD['price'].append(play.get("price"))
        


        vPD['movie_name'].append(movie.get("title"))
        vPD['movie_date'].append(movie.get("date"))
        vPD['movie_voters'].append(movie.get("voters"))
        vPD['movie_score'].append(movie.get("score"))
        vPD['movie_tags'].append(movie.get("tags"))
        vPD['movie_id'].append(movie.get("id"))
```

```python
vdf = pd.DataFrame(vPD)
```

```python
vdf.head()
```

```python
vdf.groupby("site").count().sort_values("movie_id", ascending=False)
```

```python
vdf.groupby(["site", "price"]).count()
```

```python
def get_year(date):
    if not date:
        return 0
    return date.split("-")[0]
```

```python
vdf['year'] = vdf['movie_date'].apply(get_year)
```

```python
vdf.groupby(["year","site"]).count().sort_values("movie_id", ascending=False)
```
